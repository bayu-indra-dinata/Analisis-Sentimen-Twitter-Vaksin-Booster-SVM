```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
queries = ["vaksin booster \"vaksin booster\" (vaksin OR booster) lang:id until:2022-12-31 since:2022-01-12"]
```

```python
tweets = []
max_tweets = 8000
counter = 0

for query in queries:
    query_string = query
    scraper = sntwitter.TwitterSearchScraper(query_string)
    
    for tweet in scraper.get_items():
        tweets.append([tweet.user.username, tweet.date, tweet.rawContent])
        counter += 1
        
        if counter >= max_tweets:
            break
            
    if counter >= max_tweets:
        break
```

```python
df = pd.DataFrame(tweets, columns=['user', 'date', 'renderedContent'])
```
