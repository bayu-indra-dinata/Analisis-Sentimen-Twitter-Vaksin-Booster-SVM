```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
df = pd.read_csv('dataset/tweets_labelled.csv')
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
df
```

```python
# For limit data sampel SKENARIO 1
s_1 = df[df['sentiment'] == -1].sample(400, replace=True)
s_2 = df[df['sentiment'] == 1].sample(400, replace=True)
df = pd.concat([s_1, s_2])

# Check value counts of Labels
df['sentiment'].value_counts()
```

<h1>Visualizing Word Clouds</h1>

```python
# # generate Wordcloud 

s_1 = df[df["sentiment"] == -1]
s_2 = df[df["sentiment"] == 1]

filtered_neg = " ".join(word for word in s_1["tweets"])
filtered_pos = " ".join(word for word in s_2["tweets"])
```

<h2>Wordcloud: Sentiment Negative</h2>

```python
# Polarity == -1 (negative)

wordcloud = WordCloud(colormap='Reds', margin=0, width=1500, height=1000, mode='RGBA', 
                      background_color='White', max_font_size=160).generate(filtered_neg)

# plotting wordcloud img 
plt.figure(figsize=[15, 10], facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
# plt.title("Wordcloud Sentimen Negatif")
plt.savefig('output/wordcloud_negative1.png')
plt.show()

wf = wordcloud.words_
```

<h2>Wordcloud: Sentiment Positive</h2>

```python
# Polarity == 1 (positive)

wordcloud = WordCloud(margin=0, width=1500, height=1000, mode='RGBA', 
                      background_color='white', max_font_size=160).generate(filtered_pos)

plt.figure(figsize=[15, 10], facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
# plt.title("Wordcloud Sentimen Positif")
plt.savefig('output/wordcloud_positive2.png')
plt.show()
```

<h2>Barplot Jumlah Sentimen Positif dan Negatif</h2>

Hasil jumlah sentimen dengan perbandingan data aktual <i>(labelling lexicon)</i> dengan model prediksi 4 <i>kernel</i> SVM

```python
sen_pos = (1194, 1268, 1461, 1274, 1290)
sen_neg = (583, 509, 316, 503, 487)
```

```python
xpos = np.arange(len(sen_pos))
xpos
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))
barWidth = 0.3
plt.bar(xpos, sen_pos, color='royalblue', width = barWidth, label='Positif')
plt.bar(xpos+0.3, sen_neg, color='red', width = barWidth, label='Negatif')
plt.xticks(xpos+0.2, ('Lexicon', 'Linear', 'Polynomial', 'RBF', 'Sigmoid'))
plt.xlabel('Aktual dan Model Prediksi')
plt.ylabel('Jumlah Sentimen')
# plt.title('Hasil Perbandingan Pelabelan Lexicon dengan Model Prediksi')
plt.savefig('output/barplot_lexicon_model.png')
plt.legend()
```
